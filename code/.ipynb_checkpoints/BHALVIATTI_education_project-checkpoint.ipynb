At this point I will need to accept the outstanding null data, as one data set appears to have schools combined (KENMORE/GARFIELD HIGH SCHOOL) vs having two sepearate records in our main df (GARFIELD HIGH SCHOOL and KENMORE HIGH SCHOOL). I will write a left join that first matches on id=combokey since that is the most accurate match. Then I will re-match the rows that were unmatched in my first attempt using leaid and school name as a second attempt. I will then concatenate those two dataframes into one final dataframe to get a final arrest and referral count by school.

```python
merge_attempt_1 = df.merge(
    referrals_arrest[['combokey', 'arrests', 'referrals']],
    left_on='id',
    right_on='combokey',
    how='left',
    suffixes=('', '_a1')
)

unmatched = merge_attempt_1[merge_attempt_1['arrests'].isna()].copy()

merge_attempt_2 = unmatched.merge(
    referrals_arrest[['leaid', 'sch_name', 'arrests', 'referrals']],
    on=['leaid', 'sch_name'],
    how='left',
    suffixes=('', '_a2')
)

## Using one column as a proxy for whether that row successfully matched
df_join = pd.concat([
    merge_attempt_1[merge_attempt_1['arrests'].notna()],  
    merge_attempt_2                                       
], ignore_index=True)


df_join['arrests'] = df_join['arrests'].combine_first(df_join['arrests_a2'])
df_join['referrals'] = df_join['referrals'].combine_first(df_join['referrals_a2'])

# drop the joined columns 
df_join = df_join.drop(columns=['arrests_a2', 'referrals_a2'], errors='ignore')
```

As a final check, let's verify the total rows in our dataframe, how many of them were joined successfully on combokey (attempt 1), how many were recovered in our fallback of leaid and name (attempt 2), and what amount of rows remained unmatched:

```python
total_rows = len(df)
matched_attempt1 = merge_attempt_1['arrests'].notna().sum()
matched_attempt2 = merge_attempt_2['arrests_a2'].notna().sum()
total_matched = matched_attempt1 + matched_attempt2
unmatched_remaining = total_rows - total_matched

print(f'Matches from first join (combokey=id): {matched_attempt1:,}')
print(f'Recovered from fallback (leaid+sch_name): {matched_attempt2:,}')
print(f'Still unmatched after both: {unmatched_remaining:,}')
print(f'Final dataset shape: {df_join.shape}')
```
